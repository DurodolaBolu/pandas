```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
# data.filter(['Country','Year', 'Number of reported cases of cholera', 'Number of reported deaths from cholera'])

data[['Country', 'Year', 'Number of reported cases of cholera', 'Number of reported deaths from cholera']]

data.loc[data['Country'] == 'Afghanistan'].sort_values('Year')
# data['Country'].count()

cols = list(data.columns)

data = data[cols[0:2] + [cols[-1]] + cols[2:5]]
data = data.rename(columns={'Number of reported cases of cholera': 'Cases',
                     'Cholera case fatality rate': 'Death rate(%)',
                     'Number of reported deaths from cholera': 'Deaths'})
data.dtypes
```

```python
data[data['Deaths'].str.contains('[^0-9\.]') & data['Deaths'].notnull()]
```

```python
data[data['Death rate(%)'].str.contains('[^0-9\.]') & data['Death rate(%)'].notnull()]
# Iraq has a duplicated value 
# Trying to replace the duplicated value of iran with the highest number

data.loc[1059, 'Cases'] = '5'
data.loc[1059, 'Deaths'] = '0'
data.loc[1059, 'Death rate(%)'] = '0.0'

```

```python
for column in ['Cases', 'Deaths', 'Death rate(%)']:
    data[column] = data[column].replace('Unknown', np.nan)
    data[column] = pd.to_numeric(data[column], downcast='integer')
    
data
```
