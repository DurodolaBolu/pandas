# Filtering Data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('new_meningitis.csv')
data.loc[(data['gender'] == 'Female') & ((data['age'] <= 30) | (data['state'] == 'Ogun'))]

# Using query method to perform same purpose
data.query("(gender == 'Female' & age < 30) | state == 'Ogun'")
```
